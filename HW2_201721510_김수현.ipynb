# 1. 소셜 네트워크 분석의 개요

## 1.1 네트워크 이론의 개요 

- **네트워크** : 각각의 객체들이 상호 연결된 구조
- **구성요소** : node(Node, vertex)와 링크(link, edge)
- **노드** : 고유한 속성을 가지고 있는 행위자
- **링크** : 노드들간의 관계(relation)

**즉, 네트워크는 복수의 노드들이 연결된 구조**

ex) 세계 항공기 노선도, 수도권 지하철 노선도

1. **네트워크 이론** : 그래프 이론에서 비롯됨, 현재 전산학, 생물학, 경제학, 사회학 분야에 널리 적용
2. **그래프 이론**: 점(노드)와 그 점을 잇는 선(링크)로 이뤄진 그래프를 수학적으로 연구하는 이론, 수학과 전산학분야에서 연구
3. **복잡계 네트워크 이론**: 물리학, 생물학, 경제학, 사회학 등 여러 분야에 적용

## 1.2 네트워크 이론의 역사

 1. **오일러 '쾨니히스베르크 다리건너기 문제'**
 2. **모레노 '소시오그램'**
 3. **바베라스와 리빗의 실험** : 
    + 네트워크 중심성에 대한 연구 촉발
 4. **밀그랩의 6단계 분리이론** : 
    + 좁은 세상 효과(small world effect)
 5. **그라노베터의 ‘약간 연결의 강점＇**: 
    + 약한 사회적 관계의 유용성에대한 새로운 시사점을 제시
 6. **프리만의 ‘중심성 지표‘ 개발** : 
    + 연결정도 중심성(Degree centrality), 근접 중심성(Closeness centrality), 매개 중심성(Betweenness centrality)이 네트워크 분석에 있어 가장 널리 상용되는 척도 됨
    - 매개중심성 : 매개자 역할 얼마나 수행하는지 측정 ex)배민
 7. **버트의 ‘구조적 공백‘** : 
    + 구조적 공백의 위치에 있는 노드는 다른 노드들과 중복적인 관계를 가지고 있지 않기 때문에 노드와 노드를 연결시켜 주는 역할을 함
 8. **왓츠, 스트로가츠의 ‘좁은 세상 네트워크‘** : 
    + 무작위 네트워크(Random network)의 중간에 존재
 9. **바라바시, 알버트, 정하웅의 ‘무척도 네트워크＇**: 
    + 인터넷 연결 수 분포가 정규분포를 따르지 않고 멱함수법칙(Power law)을 따르는 것을 밝혔으며 이러한 네트워크 구조를‘무작위 네트워크(Random network)’와 구별하여 ‘무척도 네트워크(Scale-free network)' 라고 명명
    + 노드 간 연결 이루어질 때 특정 노드에 대한 선호적 연결현상 발생
    +  ex)허브, 인플루언서

# 2. 네트워크 이론의 기초
## 2.1 그래프의 기본개념
- **그래프의 기본개념**
  1. **그래프이론** : 객체들 간의 네트워크 관계 모델을 수학적으로 표현된 그래프로 나타내는 연구 분야 그래프는 노드(Node, vertex) 로 구성, 이러한 노드들은 링크(Link, edge)로 연결
  2. **무방향 네트워크** : 책, 논문의 공저자, 네트워크, 페이스북의 친구 관계 네트워크
  3. **방향 네트워크** : 인터넷 주소의 하이퍼링크 네트워크, 전화통화 네트워크, 트위터의 리트윗, 팔로잉 네트워크 등
  4. **이진 네트워크** : 관계의 유무만을 표현
  5. **가중 네트워크** : 관계의 정도를 함께 표시, 관계의 정도는 링크 옆에 가중치 직접표시 or 링크 선의 굵기 다르게 표시
  
## 2.2 그래프의 표현 방법
1. **인접행렬 방식** : 인접한 노드들 간의 연결 관계를 행렬(Matrix)로 표현
    + 행과 열의 이름은 노드를 나타냄
    + 행과 열이 교차하는 지점의 행렬 값은 관계(link)의 유무를 1과 0으로 나타냄
    + 행의 이름은 링크가 출발하는 노드를 나타내고 열의 이름은 링크가 도착하는 노드를 나타냄
    + 대칭적인 특징을 가짐
2. **노드 리스트 방식** : 인접한 링크가 있는 노들만을 표기하는 방법, 노드들만의 목록
3. **엣지 리스트 방식** : 노드들을 연결한 링크들의 리스트로 “링크의 목록＂

# 3. 네트워크 유형

## 3.1 방향성과 가중치에 따른 분류
- 네트워크는 링크의 방향성 유무와 가중치 유무에 따라 크게 4가지 유형으로 구분이 가능

## 3.2 관계의 위상에 따른 분류
- 네트워크를 이루고 있는 관계 위상 형태에 따라서 구분가능
- ring, mesh, star, fully connected, line, tree, bus

## 3.3 네트워크 크기에 따른 분류
- 네트워크를 이루고 있는 노드의 링크의 규모 등 네트워크 크기에 따라 다음과 같이 구분
   1. **micro(소규모)** : 노드수 <=100 , 최대 링크 수 <= 10,000
   2. **meso(중규모)** :  노드수 <=1,000 , 최대 링크 수 <= 1,000,000
   3. **macro(대규모)** : 노드수 <=10,000 , 최대 링크 수 <= 1000,000,000
   4. **super-macro(초대규모**) :  노드수 <=10,000 , 최대 링크 수 = 무한대
   
## 3.4 분석 대상에 따른 분류
- **전체 네트워크**
    + 네트워크를 구성하고 있는 모든 노드를 포함한 네트워크를 말함. 완전 네트워크(whole network)라고도 함
- **하위 네트워크**
   + 전체 네트워크의 일부분으로 부속되어 있는 것으로 컴포넌트(component), 파당(Clique), 클러스터(Cluster) 등이 해당
- **파당(Clique)**
   + 네트워크의 하위 네트워크, 파당에 포함된 모든 노드들이 상호간에 직접적으로 연결(완전연결) 되어 있음
   + 다른 파당에는 포함되어 있지 않은 경우를 말함. 일반적으로 양자 관계는 제외, 3개 이상의 노드를 포함한 것
- **클러스터**
   + 유사한 속성(특성)을 가지고 있는 노드들의 집합
   + 유사성에 따라 클러스터를 분류해 내는 작업
   + 노드간 유사성 분석을 통한 클러스터링은 수학적 알고리즘을 기반으로 계산 및 시각화되며 대부분의 소셜네트워크 알고리즘에 내장되어 있어 손쉽게 결과물을 얻을 수 있음
- **에고 네트워크**
   + 네트워크 상에서 분석하고자 하는 특정 노드를 중심에 두고 그 노드와 직접적으로 연결된 노드들 간의 관계를 네트워크로 표현
   + 자아중심 네트워크(Ego-centric network)라고도 함

# 4. 소셜 네트워크 분석 지표

## 4.1 네트워크 기본 속성 분석

### 4.1.1 네트워크 수준의 속성 분석
- **네트워크 크기**
   + 네트워크를 구성하는 노드들의 수, 네트워크 크기가 증가할수록 네트워크 구조의 복잡성도 증가, 소셜 네트워크 분석에 있어 중요한 역할
   - **연결관계의 수**
      - 방향성이 있는 경우: nx(n-1) 
      - 방향성이 없는 경우: nx(n-1)/2
- **네트워크 밀도**
   + 네트워크를 구성하고 있는 노드들 간의 연결된 정도, 네트워크 내의 노드들 간의 연결이 많아질수록 밀도는 높아짐
   - **네트워크 밀도 계산** 
       - 무방향 이진 네트워크의 밀도: k/{n x (n-1)/2} 
       - 방향 이진 네트워크의 밀도: k/{n x (n-1)} 
       - k: 실제 연결된 링크의 수, n: 노드의 수
- **포괄성(Inclusiveness)** 
   + 전체 네트워크의 밀집도를 연결된 노드의 관점에서 표현, 네트워크상의 총 노드의 개수 중에서 실제로 연결 관계가 존재하는 노드의 개수가 차지하는 비율
   - **포괄성 계산**
      - n-nisolated 
      - n: 총 노드의 수, nisolated: 고립된 노드의 수

### 4.1.2 노드 수준의 속성 분석
- **연결정도(Degree)**
   + 해당 노드에 직접 연결되어 있는 노드들의 개수, 또는 링크의 개수 
   + n개의 노드가 존재한다고 가정 했을 때, 각각의 노드에 최대 n-1개까지의 연결 정도가 가능 
   + 고립된 노드의 경우에는 연결정도가 0 , 연결정도의 측정을 통해 전체 네트워크 구조상에서 해당 노드가 보여주는 활동성과 영향력 유추가능
   + 연결정도가 높을수록 다른 노드들과 많은 관계를 가지고 있음을 보여주며 이는 상대적으로 높은 활동성과 영향력을 반증
- **연결거리(Distance)**
   + 노드간 관계(링크)의 강도 
   + 두개의 특정 노드들 사이에 존재하는 다양한 연결경로 중에서 가장 적은 수의 링크 단계를 거치는 경로
   + 즉 가장 짧은 연결 경로, 두 노드간 연결 거리가 짧을수록 노드간의 연결이 높아짐
   + 정보나 구전 등이 얼마나 신속하게 전달되는지 가늠하는 중요한 지표
     - **이진 네트워크(Binary network)**: 관계의 유무만 표시한 네트워크
     - **가중 네트워크(Valued network)**: 연결 관계의 강도를 상대적인 가중치를 함께 표시한네트워크. 관계 빈도, 친밀성, 지속성, 신뢰성 등 여러가지 관계 특성을 정의 가능
- **직경(Diameter)** 
   + 네트워크 상의 모든 연결거리 중에서 가장 긴 연결거리
   + 네트워크 상에서 가장 멀리 떨어져있는 두개 노드의 연결거리=네트워크 지름
   + 네트워크 직경이 작을수록 신속한 정보 전달이 가능해지고 효율성 높아짐
- **평균연결거리(AVerage Distance)** 
   + 네트워크내 모든 노드 쌍의 최단 연결거리를 산출하여 평균값을 계산한것
   + 네트워크 임의의 두 노드가 평균적으로 최소 몇 단계의 링크를 걸쳐서 연결이 가능한지 보여줌

### 4.1.3 네트워크에 내재된 속성 분석
- **상호성(Reciprocity)** 
   + 방향 네트워크 상에서 2개의 노드 간의 연결이 서로 쌍방향인 경우: 상호적 연결 
   + 방향 네트워크 상에서 2개의 노드 간의 연결이 서로단방향인 경우 : 비상호적연결
   + 상호성 계산은 전체 네트워크 상에서 상호적 연결이 차지하고 있는 비율로 산출
   
- **군집화 계수(Clustering coefficient)** 
   + 군집화 계수: 군집화의 정도를 나타내는 지표
   + 해당 노드의 이웃 노드들 간에 가능한 최대 관계의 수 중에서 실제로 존재하는 관계의 수가 차지하는 비율로 정의
   + 군집화(Clustering): 네트워크 상에서 특정 부분의 노드들이 밀접하게 연결되어 다른 부분보다 네트워크 밀도가 높아지는 현상 
   

## 4.2 중심성 분석

### 4.2.1 중심성 지표
- **중심성 지표는 소셜 네트워크 분석 지표 중에서 일반적으로 가장 많이 사용되는 지표로 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표**
- 중심성 분석을 통해 전체 네트워크에서 중요한 역할을 하는 노드가 무엇인지 파악 가능
- 네트워크에서 중심적인 역할을 하는 구성원을 탐색하고 그 구성원을 통해 효율적인 정보 전달과 의사전달을 가능케함으로써 문제를 효과적으로 해결가능

### 4.2.2 연결정도 중심성
- **한 노드가 얼마나 많은 다른 노드들과 연결관계를맺고 있는 지 측정**
- 노드는 다른 노드들과 많은 연결 관계를 가질수록 더 넓은 선택의 폭과 자율성을 가지게 되어 다른 노드에 대한 의존성을 상대적으로 낮춤
- 결과적으로 네트워크 상에서 더 큰 영향력
- 노드에 연결된 링크의 수가 많아질수록 해당 노드의 연결 정도 중심성은 높아짐

### 4.2.3 근접 중심성
- **한 노드가 얼마나 전체 네트워크의 중심에 근접해 있는지를 측정**
- 해당 노드가 전체 네트워크 상에서 얼마나 중앙에 위치하고 있는지를 측정
- 노드가 중앙에 가깝게 위치할 수록 자원과 정보를 신속하게 전체 네트워크로 확산
- 근접 중심성 계산

### 4.2.4 매개 중심성
- **한 노드가 다른 노드들 간의 네트워크 관계 형성에 있어서 중개자 혹은 매개자 역할을 얼마나 수행하는 지 측정하는 지표**
- 노드의 매개 중심성이 높을 수록 네트워크 내에서 정보 및 자원의 흐름에 있어서 높은 통제력 확보
- 매개 중심성이 높은 노드는 관계 연결의 길목에서 정보를 왜곡하거나 관계 활동의 방해를 통해 네트워크를 위협할 수 있음
- 절대적 매개 중심성: 노드 i의 매개 중심성은 네트워크 내의 모든 노드 쌍의 최단 경로 중에서 노드 i가 포함되어 있는 경로의 비율로 정의됨 
- 상대적 매개중심성: 위와 같이 계산한 절대적 매개 중심성 값을 이론상 가능한 매개 중심성 최댓값으로 나누어 계산

### 4.2.5 아이겐벡터 중심성
- **연결된 다른 노드들의 중요도를 함께 반영한 것**
- 중심성이 높은 노드들과 많이 연결된 노드일수록 해당 노드의 아이겐벡터 중심성은 커짐
- 네트워크 상에서 소수 몇몇 영향력 높은 노드들에 연결되어 있는 것이 다수의 일반 노드들에 연결되어 있는 것보다 더 큰 영향력을 가짐
- 즉 친구가 많은 것도 중요하지만 중요한 친구가 많은 지를 보고자 할 때 사용하는 지표임

```python
# 모듈 가져오기
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import warnings; warnings.simplefilter('ignore')
```

## 1.1 소셜 네트워크 기본 사항
- 네트워크 구성
   + Nodes : 우리가 구축하고있는 네트워크의 개인(행위자)
   + Edges : 노드 간의 연결으로 노드 간의 관계를 나타냄

### 1.1.1  Symmetric Networks : 대칭 네트워크 (무 방향)
- 책, 논문의 공저자, 네트워크, 페이스북의 친구 관계 네트워크
- A가 B와 관련이 있으면 B도 A와 관련이 있음

```python
G_symmetric = nx.Graph() #네트워크 구성하기

G_symmetric.add_edge('Steven',  'Laura')
G_symmetric.add_edge('Steven',  'Marc')
G_symmetric.add_edge('Steven',  'John')
G_symmetric.add_edge('Steven',  'Michelle')
G_symmetric.add_edge('Laura',   'Michelle')
G_symmetric.add_edge('Michelle','Marc')
G_symmetric.add_edge('George',  'John')
G_symmetric.add_edge('George',  'Steven')
```

```python
print(nx.info(G_symmetric)) #네트워크 정보 출력
```

```python
# draw_networkx() 함수를 사용하여 네트워크 시각화
plt.figure(figsize=(5,5))
nx.draw_networkx(G_symmetric);
```

### 1.1.2 Asymmetric Networks : 비대칭 네트워크 (방향성)
- 인터넷 주소의 하이퍼링크 네트워크, 전화통화 네트워크, 트위터의 리트윗, 팔로잉 네트워크 등
-  관계가 비대칭인 네트워크 (A가 B와 관련되어 있다고해서 B가 A와 연결되어 있다는 의미는 아님)
- DiGraph method를 사용하여 NetworkX에서 비대칭 네트워크를 구축 할 수 있음

```python
#DiGraph method를 사용하여 비대칭 네트워크 구축
G_asymmetric = nx.DiGraph()
G_asymmetric.add_edge('A','B')
G_asymmetric.add_edge('A','D')
G_asymmetric.add_edge('C','A')
G_asymmetric.add_edge('D','E')
```

```python
#모든 노드가 네트워크에서 명확하게 표시되도록 하기위해 spring_layout()함수와 draw_networkx()함수를 차례로 사용
nx.spring_layout(G_asymmetric)
nx.draw_networkx(G_asymmetric)
```

### 1.1.3 Weighted Networks : 가중 네트워크
- 관계의 정도를 함께 표시, 관계의 정도는 링크 옆에 가중치 직접표시 or 링크 선의 굵기 다르게 표시

```python
G_weighted = nx.Graph()

G_weighted.add_edge('Steven',  'Laura',   weight=25) # 관계의 정도를 가중치로 표시
G_weighted.add_edge('Steven',  'Marc',    weight=8)
G_weighted.add_edge('Steven',  'John',    weight=11)
G_weighted.add_edge('Steven',  'Michelle',weight=1)
G_weighted.add_edge('Laura',   'Michelle',weight=1)
G_weighted.add_edge('Michelle','Marc',    weight=1)
G_weighted.add_edge('George',  'John',    weight=8)
G_weighted.add_edge('George',  'Steven',  weight=4)
```

```python
elarge = [(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] > 8]
esmall = [(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] <= 8]

pos = nx.circular_layout(G_weighted)  # 모든 node의 위치
# nodes
nx.draw_networkx_nodes(G_weighted, pos, node_size=700)

# edges
nx.draw_networkx_edges(G_weighted, pos, edgelist=elarge,width=6)
nx.draw_networkx_edges(G_weighted, pos, edgelist=esmall,width=6, alpha=0.5, edge_color='b', style='dashed')

# labels
nx.draw_networkx_labels(G_weighted, pos, font_size=20, font_family='sans-serif')

plt.axis('off')
plt.show();
```

## 1.2 Clustering coefficient : 군집화 계수
- 군집화 계수: 군집화의 정도를 나타내는 지표
- 해당 노드의 이웃 노드들 간에 가능한 최대 관계의 수 중에서 실제로 존재하는 관계의 수가 차지하는 비율로 정의
- 군집화(Clustering): 네트워크 상에서 특정 부분의 노드들이 밀접하게 연결되어 다른 부분보다 네트워크 밀도가 높아지는 현상
- 노드의 클러스터, **로컬 클러스터링 계수** 결정가능 -> 서로 연결된 노드의 친구(연결) 쌍의 비율

```python
nx.clustering(G_symmetric,'Michelle') #로컬 클러스터링 계수를 결정하기 위해 nx.clustering(Graph, Node)함수 사용, Michelle의 로컬 클러스터링 계수
```

```python
nx.clustering(G_symmetric,'Laura') #Laura의 로컬 클러스터링 계수
```

```python
nx.average_clustering(G_symmetric) #대칭 직원 네트워크의 평균 클러스터링 계수
```

## 1.3 Network Distance Measures : 네트워크 거리 측정
### 1.3.1 Degree : 연결정도
- 해당 노드에 직접 연결되어 있는 노드들의 개수, 또는 링크의 개수 
-  n개의 노드가 존재한다고 가정 했을 때, 각각의 노드에 최대 n-1개까지의 연결 정도가 가능 
- 고립된 노드의 경우에는 연결정도가 0 , 연결정도의 측정을 통해 전체 네트워크 구조상에서 해당 노드가 보여주는 활동성과 영향력 유추가능
- 연결정도가 높을수록 다른 노드들과 많은 관계를 가지고 있음을 보여주며 이는 상대적으로 높은 활동성과 영향력을 반증

```python
nx.degree(G_symmetric, 'Michelle') # Michelle이 네트워크에서 3 명의 직원과 함께 작업 했으므로이 값은 3을 반환
```

### 1.3.2 Distance : 연결거리
- 노드간 관계(링크)의 강도 
- 두개의 특정 노드들 사이에 존재하는 다양한 연결경로 중에서 가장 적은 수의 링크 단계를 거치는 경로
- 즉 가장 짧은 연결 경로, 두 노드간 연결 거리가 짧을수록 노드간의 연결이 높아짐
- 정보나 구전 등이 얼마나 신속하게 전달되는지 가늠하는 중요한 지표
  - **이진 네트워크(Binary network)**: 관계의 유무만 표시한 네트워크
  - **가중 네트워크(Valued network)**: 연결 관계의 강도를 상대적인 가중치를 함께 표시한네트워크. 관계 빈도, 친밀성, 지속성, 신뢰성 등 여러가지 관계 특성을 정의 가능

```python
nx.shortest_path(G_symmetric, 'Michelle', 'John') #nx.shortest_path(Graph, Node1, Node2)함수 사용해 두 노드 간 최단경로와 길이 결정
```

```python
nx.shortest_path_length(G_symmetric, 'Michelle', 'John') #nx.shortest_path_length(Graph, Node1, Node2)함수 사용해 두 노드 간 최단경로와 길이 결정
```

### 1.3.3 Breadth-first search: 폭 우선 검색
- 해당 노드에서 시작하여 Breadth-first search 알고리즘을 사용해 노드에서 네트워크의 다른 노드까지의 거리를 찾을 수 있음

```python
S = nx.bfs_tree(G_symmetric, 'Steven') #bfs_tree 함수 사용해 노드간의 거리 구함
nx.draw_networkx(S)
```

```python
M = nx.bfs_tree(G_symmetric, 'Michelle') #M = nx.bfs_tree(G_symmetric, 'Michelle')함수를 사용해 michelle에서 시작해 네트워크의 다른 노드에 도달하는 방법 알려주는 네트워크 구조 확인
nx.draw_networkx(M)
```

## 1.4 Centrality measures: 중심성 측정
- Centrality Measures 는 네트워크 내에서 가장 영향력있는 사람을 식별하는 데 도움됨

### 1.4.1 Degree Centrality: 연결정도 중심성
- 한 노드가 얼마나 많은 다른 노드들과 연결관계를맺고 있는 지 측정
- 노드는 다른 노드들과 많은 연결 관계를 가질수록 더 넓은 선택의 폭과 자율성을 가지게 되어 다른 노드에 대한 의존성을 상대적으로 낮춤
- 결과적으로 네트워크 상에서 더 큰 영향력
- 노드에 연결된 링크의 수가 많아질수록 해당 노드의 연결 정도 중심성은 높아짐

```python
nx.degree_centrality(G_symmetric) #degree_centrality()함수를 사용해 중심성 계산
```

### 1.4.2 Eigenvector Centrality: 고유 벡터 중심성
- 얼마나 많은 개인이 연결되어 있는지뿐만 아니라 연결되어있는 사람들의 유형이 노드의 중요성을 결정
- 고유 벡터 중심성은 노드가 다른 중요한 노드에 얼마나 잘 연결되어 있는지를 고려하여 노드를 가져 오는 방법을 측정

```python
nx.eigenvector_centrality(G_symmetric) #eigenvector_centrality()함수를 사용해 고유 벡터 중심성 계산
```

### 1.4.3 Closeness Centrality:  근접 중심성
- Closeness Centrality는 각 노드의 중요도가 다른 모든 노드에 대한 근접성에 의해 결정되는 척도
- 한 노드가 얼마나 전체 네트워크의 중심에 근접해 있는지를 측정
- 해당 노드가 전체 네트워크 상에서 얼마나 중앙에 위치하고 있는지를 측정
- 노드가 중앙에 가깝게 위치할 수록 자원과 정보를 신속하게 전체 네트워크로 확산


```python
nx.closeness_centrality(G_symmetric) #Closeness Centrality()함수를 사용해 근접성 계산
```

### 1.4.4 Betweenness Centrality:  사이의 중심성
- Betweenness Centrality는 제어의 중심
- 포인트 쌍을 연결 한 최단 경로 에서 포인트가 발생하는 빈도를 나타냄 & 특정 노드가 다른 두 노드 사이에서 선택한 최단 경로로 오는 횟수를 수량화함
- 중간 중심성이 높은 노드는 네트워크 내의 통신 / 정보 흐름에서 중요한 역할 함
- 중간 중심성이 높은 노드는 전략적으로 제어하고 다른 노드에 영향을 미칠 수 있음, 이러한 위치에있는 개인은 전송되는 정보를 보류하거나 채색함으로써 전체 그룹에 영향을 미칠 수 있음
- betweenness_centrality()로 네트워크에 대해 측정가능

```python
nx.betweenness_centrality(G_symmetric) #betweenness_centrality()함수를 사용해 중심성 측정
```

```python
pos = nx.spring_layout(G_symmetric)
betCent = nx.betweenness_centrality(G_symmetric, normalized=True, endpoints=True) #사이 값을 정규화할지 여부, 가중치를 중심성 계산에 포함할지 여부, 최단 경로 수에 끝점을 포함할지 여부 선택
node_color = [20000.0 * G_symmetric.degree(v) for v in G_symmetric]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_symmetric, pos=pos, with_labels=True,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```

```python
sorted(betCent, key=betCent.get, reverse=True)[:5] #중간 중심성이 높은 노드 순으로 정렬
```

```python
# Betweenness Centrality를 사용하여 노드 색상이 Degree 및 노드 크기에 따라 달라 지도록 네트워크를 시각화
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```

```python
# betweenness centrality가 가장 높은 노드의 레이블을 알 수 있음
sorted(betCent, key=betCent.get, reverse=True)[:5]
```
